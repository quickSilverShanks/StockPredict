```python
import pandas as pd
import numpy as np
import os
import re
import calendar
import datetime
```

```python
n50_lookup = pd.read_csv("nifty50_lookuptable.csv")
n50_lookup.head()
```

```python
def get_filenames(i):
    '''
    Get available pickle dataset names for the company in row 'i' of lookup table 'n50_lookup'
    '''
    
    global filenames_ystk, filenames_mctrl, filenames_hndu, ticker

    thehindu_searchstring, ticker, yfin_ticker = n50_lookup[['thehindu_searchstring', 'ticker', 'yfin_ticker']].loc[i]

    filenames_ystk = ' '.join(os.listdir("../stocks_yahoo"))
    rgx = '(stocks_yfinance_' + yfin_ticker + '_\S*).pkl'
    filenames_ystk = re.findall(rgx, filenames_ystk)

    filenames_mctrl = ' '.join(os.listdir("../news_moneycontrol"))
    rgx = '(news_mcontrol_' + ticker + '_\S*).pkl'
    filenames_mctrl = re.findall(rgx, filenames_mctrl)

    filenames_hndu = ' '.join(os.listdir("../news_thehindu"))
    rgx1 = '(news_thehindu_' + thehindu_searchstring.replace('%20','_') + ').pkl'
    rgx2 = '(news_thehindu_' + thehindu_searchstring.replace('%20','_') + '_id\S*).pkl'
    filenames_hndu = re.findall(rgx1, filenames_hndu) + re.findall(rgx2, filenames_hndu)

# Please note that this function has fixed the filename formats in which various extracted datasets will be saved.
# This filename format must be followed to avoid discrepancy.
```

```python
def get_totaldata():
    '''
    Append all the available datasets for each company
    '''
    
    data_mctrl = pd.DataFrame()
    for data in filenames_mctrl:
        print("Import for", data, " :", end=' ')
        dset = pd.read_pickle("../news_moneycontrol/"+data+".pkl")
        print(dset.shape)
        data_mctrl = pd.concat([data_mctrl, dset], ignore_index=True)
    print("Total data(with duplicates) from moneycontrol for", ticker, " :", data_mctrl.shape)
    data_mctrl.drop_duplicates()
    print("Total usable data from moneycontrol for", ticker, " :", data_mctrl.shape)
    data_mctrl.to_pickle("news_moneycontrol_"+ticker+".pkl")

    data_ystk = pd.DataFrame()
    for data in filenames_ystk:
        print("Import for", data, " :", end=' ')
        dset = pd.read_pickle("../stocks_yahoo/"+data+".pkl")
        print(dset.shape)
        data_ystk = pd.concat([data_ystk, dset], ignore_index=True)
    print("Total data(with duplicates) from yahoo for", ticker, " :", data_ystk.shape)
    data_ystk.drop_duplicates()
    print("Total usable data from yahoo for", ticker, " :", data_ystk.shape)
    data_ystk.to_pickle("stocks_yahoo_"+ticker+".pkl")

    data_hndu = pd.DataFrame()
    for data in filenames_hndu:
        print("Import for", data, " :", end=' ')
        dset = pd.read_pickle("../news_thehindu/"+data+".pkl")
        print(dset.shape)
        data_hndu = pd.concat([data_hndu, dset], ignore_index=True)
    print("Total data(with duplicates) from thehindu for", ticker, " :", data_hndu.shape)
    data_hndu.drop_duplicates()
    print("Total usable data from thehindu for", ticker, " :", data_hndu.shape)
    data_hndu.to_pickle("news_thehindu_"+ticker+".pkl")
```

```python
for i in range(50):
    get_filenames(i)
    get_totaldata()
```

```python
def inp_days(inp_data, n):
    '''
    Prepare a output data from daily stock prices ('inp_data') with a date 'n' days prior to stock price.
    '''
    out_data = inp_data.copy()
    out_data['input_day'] = n
    out_data['Published_Dates'] = out_data['Date'] - datetime.timedelta(days=n)
    return out_data
```

```python
def base1_prep(tikr):
    '''
    Prepare a base with past and future stock prices and target variables.
    '''
    
    stkprice = pd.read_pickle("stocks_yahoo_"+tikr+".pkl")
    
    # temp3 = temp3.set_index('Date').asfreq('D').reset_index()

    stkprice['Day'] = stkprice['Date'].apply(lambda x : x.day_name())
    stkprice['close_d1'] = stkprice['Close'].shift(-1)
    stkprice['close_d2'] = stkprice['Close'].shift(-2)
    stkprice['close_d3'] = stkprice['Close'].shift(-3)
    stkprice['close_d4'] = stkprice['Close'].shift(-4)

    stkprice['change_d1'] = stkprice['close_d1'] - stkprice['Close']
    stkprice['change_d2'] = stkprice['close_d2'] - stkprice['Close']
    stkprice['change_d3'] = stkprice['close_d3'] - stkprice['Close']
    stkprice['change_d4'] = stkprice['close_d4'] - stkprice['Close']

    stkprice['target_d1'] = stkprice['change_d1'].apply(lambda x : np.NaN if pd.isna(x) else (0 if x<0 else 1))
    stkprice['target_d2'] = stkprice['change_d2'].apply(lambda x : np.NaN if pd.isna(x) else (0 if x<0 else 1))
    stkprice['target_d3'] = stkprice['change_d3'].apply(lambda x : np.NaN if pd.isna(x) else (0 if x<0 else 1))
    stkprice['target_d4'] = stkprice['change_d4'].apply(lambda x : np.NaN if pd.isna(x) else (0 if x<0 else 1))
    
    base1 = pd.concat([inp_days(stkprice, 0), inp_days(stkprice, 1), inp_days(stkprice, 2), inp_days(stkprice, 3),
                       inp_days(stkprice, 4), inp_days(stkprice, 5), inp_days(stkprice, 6)],
                      ignore_index=True)
    return base1
```

```python
def create_NLPbase(tikr):
    '''
    This function uses the stock price details and articles to create a base dataset for NLP model training.
    '''
    
    base2 = pd.read_pickle("news_moneycontrol_"+tikr+".pkl").drop(['Source_URLs'], axis=1).astype({'Published_Dates': 'datetime64[ns]'})
    base3 = pd.read_pickle("news_thehindu_"+tikr+".pkl").drop(['Source_URLs'], axis=1).astype({'Published_Dates': 'datetime64[ns]'})
    base1 = base1_prep(tikr)
    
    print("Data import (MoneyControl) for", tikr, ":", base2.shape)
    print("Data import (TheHindu) for", tikr, ":", base3.shape)
    
    merged_data1 = pd.merge(base1,
                            base2,
                            on='Published_Dates',
                            how='inner')
    merged_data1['Portal'] = 'MoneyControl'
    print("NLP Base (MoneyControl) for", tikr, ":", merged_data1.shape)
    
    merged_data2 = pd.merge(base1,
                            base3,
                            on='Published_Dates',
                            how='inner')
    merged_data2['Portal'] = 'TheHindu'
    print("NLP Base (TheHindu) for", tikr, ":", merged_data2.shape)
    
    return pd.concat([merged_data1, merged_data2], ignore_index=True)
```

```python
data_nlpbase = pd.DataFrame()

for i in range(50):
    tikr = n50_lookup['ticker'][i]
    dset = create_NLPbase(tikr)
    data_nlpbase = pd.concat([data_nlpbase, dset], ignore_index=True)
    print("NLP Base updated with data for", tikr, ":", data_nlpbase.shape)
```
